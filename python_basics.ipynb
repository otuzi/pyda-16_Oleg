## Задание 1

Даны 2 переменных, в которых хранятся строки произвольной длины: *phrase_1* и *phrase_2*.  
Напишите код, который проверяет какая из этих строк длиннее.

```

```

Результат:

```

phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

3.     

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:

````python
a = 'Насколько проще было бы писать программы, если бы не заказчики'
b = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'

phrase_1 = a
phrase_2 = b

if (len (phrase_1) > len (phrase_2)): 
        print ('Фраза 1 длиннее фразы 2')

print('Конец программы')
````

Результат:  
`Ваш знак зодиака: Скорпион` 


````python
user_input_day = int(input('Введите день, (в формате xx, от 01 - 31): '))
user_input_month = str(input('Введите месяц, (пример: январь, февраль и т.д.): '))

if ((user_input_month == 'март') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'апрель') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Овен') 
elif ((user_input_month == 'апрель') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'май') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Телец') 
elif ((user_input_month == 'май') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'июнь') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Близнецы') 
elif ((user_input_month == 'июнь') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'июль') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Рак') 
elif ((user_input_month == 'июль') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'август') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Лев') 
elif ((user_input_month == 'август') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'сентябрь') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Дева')
elif ((user_input_month == 'сентябрь') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'октябрь') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Весы')
elif ((user_input_month == 'октябрь') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'ноябрь') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Скорпион')
elif ((user_input_month == 'ноябрь') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'декабрь') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Стрелец')
elif ((user_input_month == 'декабрь') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'январь') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Козерог')
elif ((user_input_month == 'январь') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'февраль') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Водолей')
elif ((user_input_month == 'февраль') and (21 <= user_input_day <= 31) or 
    (user_input_month == 'март') and (1 <= user_input_day <= 20)): print ('Ваш знак зодиака: Рыбы')

    
else: print ('Вы ввели дату не в верном формате')

print('Конец программы')
````
